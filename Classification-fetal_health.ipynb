# Splitting the data into X and y

```python
X = data.drop("fetal_health", axis =1)
y = data.fetal_health
```

```python
X.head()
```

```python
y.head()
```

# Splitting the data into training and Testing

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state=101)
```

# Standardization of the data

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler()
X_train_scaled = sc.fit_transform(X_train)
X_test_scaled = sc.transform(X_test)
```

```python
import warnings
from sklearn.exceptions import DataConversionWarning
warnings.filterwarnings(action='ignore', category=DataConversionWarning)
```

# Model Building

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import GridSearchCV
```

# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
params = {'penalty' : ['l1','l2','elasticnet'],
            'tol' : [0.0001, 0.001, 0.01,0.1],
            'C': [0.1, 1, 10, 100],
            'solver': ['newton-cg','lbfgs','liblinear','sag','saga'],
            'max_iter': [1000,1500,2000]
         }

log = GridSearchCV(estimator=model, param_grid=params, n_jobs=-1, cv=5, scoring='accuracy',error_score=0)

```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(LogisticRegression(C = 100, penalty = 'l1',solver = 'liblinear',max_iter=1000,tol=0.01),
                X,y,cv=5).mean()
```

```python
log.fit(X_train_scaled, y_train)
```

```python
log.best_params_
```

```python
log.best_score_
```

```python
log.score(X_test_scaled,y_test)
```

```python
log.score(X_train_scaled,y_train)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix,accuracy_score
logmodel = LogisticRegression(solver = 'saga', max_iter=1000, tol=0.001, random_state = 0)
logmodel.fit(X_train_scaled, y_train)
predictions = logmodel.predict(X_test_scaled)

target_names = ['Normal', 'Suspect', 'Pathological']

print(classification_report(y_test, predictions,target_names=target_names))
print(confusion_matrix(y_test, predictions))
print(accuracy_score(y_test, predictions))
```

#  Combined code for Hyper Parameter Tuning in all the models Together

```python
# import warnings filter
from warnings import simplefilter
# ignore all future warnings
simplefilter(action='ignore', category=FutureWarning)
```

```python
model_param = {
    'LogisticRegression' : {
        'model': LogisticRegression(),
        'param':{
            'penalty' : ['l1','l2','elasticnet','none'],
            'tol' : [0.0001, 0.001, 0.01,0.1],
            'C': [0.1, 1, 10, 100],
            'solver': ['newton-cg','lbfgs','liblinear','sag','saga']
            
        }                      
       
    },
          
      'DecisionTreeClassifier':{
        'model':DecisionTreeClassifier(),
        'param':{
            'criterion': ['gini','entropy'],
            'splitter':['best', 'random'],'max_features':['auto', 'sqrt', 'log2']
            
        }
    },
        'KNeighborsClassifier':{
        'model':KNeighborsClassifier(),
        'param':{
            'n_neighbors': [5,10,15,20,25]
            ,'weights':['uniform', 'distance'],'algorithm':['auto', 'ball_tree', 'kd_tree', 'brute'],
            'leaf_size':[10,20,30,40,50]
        }
    },
    
    'RandomForestClassifier':{
        'model' : RandomForestClassifier(),
        'param': {
            'n_estimators':[20,40,60,80,100],
            'criterion' : ['gini','entropy'],
            'max_depth':[2,4,6,8],
            'max_features':['auto','sqrt','log2']
        }
    },
    
    'AdaBoostClassifier' :{
        'model': AdaBoostClassifier(),
        'param': {
            'base_estimator': [DecisionTreeClassifier(max_depth=1),
                            DecisionTreeClassifier(max_depth=5),
                            DecisionTreeClassifier(max_depth=10)],
            'n_estimators':[20,40,60,80,100]
        }
    },
    

        'SVC':{
        'model':SVC(),
        'param':{
            'kernel':['rbf','linear','poly','sigmoid'],
            'C': [0.1, 1, 10, 100],
            'degree': [2,3],
            'gamma' : [0.03, 0.3, 1]
         
        }
    }
    
}
```

```python
scores =[]
for model_name, mp in model_param.items():
    model_selection = GridSearchCV(estimator=mp['model'],param_grid=mp['param'],cv=5,return_train_score=False)
    model_selection.fit(X_train_scaled,y_train)
    scores.append({
        'model': model_name,
        'best_score': model_selection.best_score_,
        'best_params': model_selection.best_params_
    })
```

```python
data_model_score = pd.DataFrame(scores,columns=['model','best_score','best_params'])
pd.set_option('display.max_colwidth', -1)
data_model_score
```
